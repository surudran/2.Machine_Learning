```python
import pandas as pd # All the funcions in the pandas will be loaded to memory, pandas will be called as pd
```

```python
dataset = pd.read_csv("Salary_Data.csv")  # read_csv funcion will be called from class pd to read the file
```

```python
dataset  #YearsExperience is input and Salary is output.
```

```python
independent = dataset[["YearsExperience"]] #independent as input, take the input column (YearsExperience) separately and assigned to variable independent.
independent
```

```python
dependent = dataset[["Salary"]]  #dependent as output, take the output column (Salary) separately and assinged to variable dependent
dependent
```

```python
from sklearn.model_selection import train_test_split # import specific function train_test_spilit the function from sklearn.model_selection
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0) # split the dataset to train set for 70% and test set for 30% assigned to variables X_train, X_test, y_train, y_test respectively
# need to divide into X_train, X_test, y_train, y_test to use the sklearn model
```

```python
from sklearn.linear_model import LinearRegression  # Procedure to import specific funtion LinearRegression from sklearn.linear_model
regressor = LinearRegression() # Procedutre for all the functions from LinearRegression() class will be assigned to model variable regressor
regressor.fit(X_train,y_train) # fit functions will be called from LinearRegression() class by providing the train set of X and y (y=WX+b).  It will calculate weight and bias using the X_train and y_train.
```

```python
weight = regressor.coef_ # coef_ method will call the weight and will be assigned to variable weight.
weight
```

```python
bais = regressor.intercept_  # intercept_ method will call the intercept and will be assigned to variable bais
bais
```

```python
y_pred = regressor.predict(X_test)  # Predicted values of y will be calculated by providing X_test values.
```

```python
from sklearn.metrics import r2_score # to find the r_score value import the function r2_score
r_score = r2_score(y_test, y_pred) # Calculate r2_score value using the y_test values and y_pred values.
```

```python
r_score  # if r_score is nearly 1, model is good.
```

```python
import pickle  # Import pickle function to save the model.
filename = "finalized_model_linear.sav"  # to save the model to file name finalized_model_linear.sav
```

```python
pickle.dump(regressor, open(filename, 'wb'))  # Save the model to filname  wb: write binary
```

```python
loaded_model = pickle.load(open("finalized_model_linear.sav", 'rb'))  # Load (read) the model from saved filename "finalized_model_linear.sav" and assinged to variable loaded_model
result = loaded_model.predict([[13]])   # Find the predicted value for given input value 15 and assigned to variable result.
```
