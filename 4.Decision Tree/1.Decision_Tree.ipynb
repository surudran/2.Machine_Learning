```python
#pandas is a library and it contains the functions related to the database.
import pandas as pd  #All the funcions in the pandas will be loaded to memory, pandas will be called as pd
```

```python
##read_csv function called to read the csv file and store to dataset.
dataset = pd.read_csv("50_Startups.csv")
```

```python
##values in the dataset will be displayed
dataset
```

```python
#Convert categarical data to numerical data using the function get_dummies, drop_first will delete the first column
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#display the values in the dataset
dataset
```

```python
#display columns in the dataset
dataset.columns
```

```python
#Spliting input values independent as input, take the input column (YearsExperience) separately and assigned to variable independent.
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
#displays the input values from variable independent
independent
```

```python
# Spliting output values, dependent as output, take the output column (Profit) separately and assinged to variable dependent
dependent = dataset[['Profit']]
```

```python
#The sklearn. model_selection module in Scikit-Learn provides functions for splitting data into training and test sets, evaluating machine learning models, and performing cross-validation.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#Procedure to import specific function DecisionTreeRegressor from sklearn.tree to create model
from sklearn.tree import DecisionTreeRegressor
#criterion{“squared_error”, “friedman_mse”, “absolute_error”, “poisson”}, default=”squared_error” , splitter{“best”, “random”}, default=”best”
regressor = DecisionTreeRegressor(criterion = 'poisson', splitter ='random') 
regressor = regressor.fit(X_train, y_train) #once excuted this line, model created and saved to variable regressor
```

```python
#Plot graph using the given data from regressor
import matplotlib.pyplot as plt  # this library for graph
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Predicted values of y will be calculated by providing X_test values.
y_pred = regressor.predict(X_test)
```

```python
#to find the r_score value import the function r2_score
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
##if r_score is nearly 1, model is good.
r_score
```

```python
#Import pickle function to save the model.
import pickle
filename = "finalized_model_DeccisionTree.sav" #to save the model to file name finalized_model_DeccisionTree.sav
pickle.dump(regressor,open(filename, 'wb'))  #Save the model to filname  wb: write binary
```
