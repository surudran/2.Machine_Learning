```python
#pandas is a library and it contains the functions related to the database.
import pandas as pd  #All the funcions in the pandas will be loaded to memory, pandas will be called as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
##read_csv function called to read the csv file and store to dataset.
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
##values in the dataset will be displayed
dataset
```

```python
#Convert categarical data to numerical data using the function get_dummies, drop_first will delete the first column
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#display the values in the dataset
dataset
```

```python
dataset=dataset.drop("User ID",axis=1)
```

```python
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
#Spliting input values and output values
indep=dataset[["Age","EstimatedSalary","Gender_Male"]]
dep=dataset["Purchased"]
```

```python
indep.shape
```

```python
dep
```

```python
#split into training set and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
#Procedure to import specific function RandomForestClassifier from sklearn.ensemble to create model
from sklearn.ensemble import RandomForestClassifier
#n_estimatorsint, default=100, criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”}, default=”squared_error”
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0) 
classifier.fit(X_train, y_train) #once excuted this line, model created and saved to variable classifier
```

```python
#Predicted values of y will be calculated by providing X_test values.
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
```

```python
print(clf_report)
```
