```python
#pandas is a library and it contains the functions related to the database
#all the functions in the pandas will be loaded to memory, pandas will be called as pd
import pandas as pd 
```

```python
#read_csv function called to read the csv file and store to dataset
dataset = pd.read_csv("insurance_pre.csv")
```

```python
#display values from dataset
dataset
```

```python
#convert categarical data to numerical data using the function get_dummies, drop_first will delete the first column
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#display the values from dataset
dataset
```

```python
#display columns from dataset
dataset.columns
```

```python
#splitting input values into independent as input
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
#display input values from varialbe independent
independent
```

```python
#splitting output values into variable dependent
dependent = dataset[['charges']]
```

```python
#display the output values from dependent
dependent
```

```python
#splitting the data into training and test sets using sklearn model selection
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#display X_train values
X_train
```

```python
#Standardisation, calculate mean and deviation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#display standardisation values
X_train
```

```python
#procedure to import SVR function from SVM and same model
from sklearn.svm import SVR
regressor = SVR(kernel = "rbf", C=3000)
regressor.fit(X_train, y_train)
```

```python
#find the intercept value
regressor.intercept_
```

```python
#check how many support vector taken to create model
regressor.n_support_
```

```python
#display support vector
regressor.support_
```

```python
#calculate predict y values using X test values
y_pred = regressor.predict(X_test)
```

```python
#find the r_score value
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
#display r_score value
r_score
```

```python
#Import pickle function to save the model.
import pickle
filename = "finalized_model_Regression_Assignment.sav"
pickle.dump(regressor, open(filename, 'wb'))
```
